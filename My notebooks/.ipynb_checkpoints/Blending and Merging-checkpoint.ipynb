## Color Model Conversion

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/00-puppy.jpg')
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Conversion from RGB to HSL

```python
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
```

### Conversion from RGB to HSL

```python
img_hls = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HLS)
plt.imshow(img_hls)
```

## Image Stitching / Blending using OpenCV

```python
img1 = cv2.imread('DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure(0)
plt.imshow(img1)
plt.figure(1)
plt.imshow(img2)
```

```python
print(img1.shape)
print(img2.shape)
```

### 1. Blending equally sized images

```python
img1 = cv2.resize(img1, (1200, 1200))
img2 = cv2.resize(img2, (1200, 1200))

plt.figure(0)
plt.imshow(img1)
plt.figure(1)
plt.imshow(img2)
```

```python
img_blended = cv2.addWeighted(src1 = img1, alpha = 0.8, src2 = img2, beta = 0.2, gamma = 0)
plt.imshow(img_blended)
```

### 2. Resizing and Overlaying images of different sizes

```python
# Overlay a smaller image on top of a larger image(No blending)
# Numpy assignment

img1 = cv2.imread('DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img2 = cv2.resize(img2, (600, 600))
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[1]
```

```python
large_img[y_offset:y_end, x_offset:x_end] = small_img
plt.imshow(large_img)
```

### 3. Resizing and Blending images of different sizes

```python
img1 = cv2.imread('DATA/dog_backpack.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (600, 600))
plt.imshow(img2)
```

```python
plt.imshow(img1)
print(img1.shape)
```

```python
# Create an ROI on the larger image where the smaller image is to be blent in
x_offset = 100
y_offset = 300

rows, columns, channels = img2.shape

x_end = x_offset + columns
y_end = y_offset + rows

roi = img1[y_offset:y_end, x_offset:x_end]

plt.imshow(roi)
```

```python
img2_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2_gray, cmap='gray')
```

```python
# Invert the color to perform the masking
mask_inv = cv2.bitwise_not(img2_gray)
plt.imshow(mask_inv, cmap='gray')
```

```python
import numpy as np

# Add the color channels so that the 2 arrays to blend have the same size
# np.full fills the given shape with the given values
white_background = np.full(img2.shape, 255, dtype = np.uint8)
white_background.shape
```

```python
background = cv2.bitwise_or(white_background, white_background, mask = mask_inv)
background.shape
```

```python
plt.imshow(background)
```

```python
foreground = cv2.bitwise_or(img2, img2, mask = mask_inv)
plt.imshow(foreground)
```

```python
final_roi = cv2.bitwise_or(roi, foreground)
plt.imshow(final_roi)
print(final_roi.shape)
```

```python
large_img = img1
small_img = final_roi

print(x_offset)
print(x_end)
print(y_offset)
print(y_end)

large_img[y_offset:y_end, x_offset:x_end] = small_img

plt.imshow(large_img)
```
