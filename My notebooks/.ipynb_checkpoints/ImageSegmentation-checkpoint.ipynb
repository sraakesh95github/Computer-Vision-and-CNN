## Watershed algorithm

```python
img = cv2.imread("../DATA/pennies.jpg")
```

```python
# 35 represents the kernel size for the blur. 35 is huge as the image is huge
img = cv2.medianBlur(img, 35)
display(img)
```

```python
gray_coins = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray_coins, 127, 255, cv2.THRESH_BINARY_INV)
display(thresh)
```

```python
# There are some artefacts in the image. To remove those apply the 
# Otsu's method - a method of thresholding that works really well in conjuncture with Watershed
ret, otsu_thresh = cv2.threshold(gray_coins, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
display(otsu_thresh)
```

```python
import numpy as np

# Perform segmentation to separate out the coins
# Perform noise removal
kernel = np.ones((3,3), np.uint8)

# Ref for morphological operations: https://docs.opencv.org/trunk/d9/d61/tutorial_py_morphological_ops.html
opening = cv2.morphologyEx(otsu_thresh, cv2.MORPH_OPEN, kernel, iterations = 2)
display(opening)
```

```python
# Set seeds for the Watershed algorithm to ensure that we stick to the foreground
# One seed would be required for every object to be identified separately
# on an image

# Perform a DISTANCE TRANSFORM
dist_transform_coins = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
display(dist_transform_coins)
```

```python
# Apply threshold to get the 6 dots and use the dots as the seeds for
# Watershed algorithm

ret, sure_fg = cv2.threshold(dist_transform_coins, 0.7 * dist_transform_coins.max(), 255, 0)
display(sure_fg)
```

```python
# Get the inknown regions
sure_fg = np.uint8(sure_fg)
sure_bg = opening.copy()
unknown = cv2.subtract(sure_bg, sure_fg)
display(unknown)
```

```python
# Create the markers for the Watershed algo
ret, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1
```

```python
# Apply the Watershed algo to the markers
markers = cv2.watershed(img, markers)
display(markers)
```

```python
contours, hierarchy = cv2.findContours(markers.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    if(hierarchy[0][i][3] == -1):
        cv2.drawContours(sep_coins, contours, i, (255,0,0), 10)
        
display(sep_coins)
```
