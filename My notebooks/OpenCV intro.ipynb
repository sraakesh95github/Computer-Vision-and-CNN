### Numpy and OpenCV in image processing

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pic = Image.open('DATA/00-puppy.jpg')
type(pic)
pic_arr = np.asarray(pic)
print(pic_arr.shape)
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
# The 3D array represents the image dimensions as (row, column, color)
pic_red = pic_red[:,:,0]
plt.figure(0)
plt.title("red image")
plt.imshow(pic_red, cmap = 'gray')
pic_green = pic_arr[:,:,1]
plt.figure(1)
plt.title("green image")
plt.imshow(pic_green, cmap = 'gray')
pic_blue = pic_arr[:,:,2]
plt.figure(2)
plt.title("blue image")
plt.imshow(pic_blue, cmap = 'gray')
```

### Creating a dual color channel image

```python
pic_rb = pic_arr.copy()
pic_gb = pic_arr.copy()
pic_rg = pic_arr.copy()
pic_rb[:,:,1] = 0
pic_gb[:,:,0] = 0
pic_rg[:,:,2] = 0

plt.figure(3)
plt.imshow(pic_rb)
plt.figure(4)
plt.imshow(pic_gb)
plt.figure(5)
plt.imshow(pic_rg)
```
