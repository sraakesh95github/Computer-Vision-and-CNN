# Image Segmentation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Watershed Algorithm

```python
def disp(img, img_type = 'gray'):
    fig = plt.figure(figsize = (12,12))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = img_type)
```

```python
coins = cv2.imread("../DATA/pennies.jpg")
disp(coins)
```

```python
#Output must be an image with 7 segments including the background

# Apply a median blur
# Gets rid of the features like the faces on the coins
# Binary threshold
# Get the contours
```

### Apply a median blur

```python
coin_blur = cv2.medianBlur(coins, 25)
disp(coin_blur)
```

### Convert to grayscale

```python
gray_coins = cv2.cvtColor(coin_blur, cv2.COLOR_BGR2GRAY)
disp(gray_coins)
```
