## Image Thresholding

### 1. Binary Thresholding

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
img=cv2.imread('DATA/rainbow.jpg', 0)
plt.imshow(img, cmap = 'gray')
```

```python
# Perform a threshold
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
print(ret)
plt.imshow(thresh1, cmap = 'gray')
```

```python
ret, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh2, cmap= 'gray')
```

```python
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(thresh3, cmap= 'gray')
```

```python
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
plt.imshow(thresh3, cmap= 'gray')
```

```python
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(thresh3, cmap= 'gray')
```

#### 1. a. Real-world application for thresholding

```python
img2 = cv2.imread('DATA/crossword.jpg', 0)
plt.imshow(img2, cmap = 'gray')
```

```python
# Resize the image window for display
def show_pic(img):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    
show_pic(img2)
```

```python
ret, img_bin = cv2.threshold(img2, 190, 255, cv2.THRESH_BINARY)
show_pic(img_bin)
```

#### 1. b. Adaptive binary threshold

```python
# This is done to get better results on the above example
thresh3 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)
show_pic(thresh3)
```

```python
blended = cv2.addWeighted(src1 = thresh3, alpha = 0.2, src2 = img_bin, beta = 0.8, gamma = 0)
show_pic(blended)
```
