# Counter Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def disp(img, type = 'g'):
    fig = plt.figure(figsize = (12, 12))
    ax = fig.add_subplot()
    if(type == 'g'):
        ax.imshow(img, cmap = 'gray')
    else:
        ax.imshow(img)
```

```python
img = cv2.imread('../DATA/internal_external.png', 0)
plt.imshow(img, cmap='gray')
print(img.shape)
```

```python
img_cpy = img.copy()

# Extraxt both internal and external contours
contours, hierarchy = cv2.findContours(img_cpy, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
print(type(contours))
print(len(contours))
print(type(hierarchy))
print(hierarchy)
```

```python
external_contours = np.zeros(img.shape)
print(external_contours.shape)
```

```python
plt.imshow(img_cpy)
```

### 1. Get the external contours

```python
external_contours = np.zeros(img.shape)

# Get the contours to be plotted
for i in range(len(contours)):
    
    # Get external contours 
    if(hierarchy[0][i][3] == -1):
        cv2.drawContours(external_contours, contours, i, 255, 2)
        
disp(external_contours)
```

### 2. Get the intenal contours

```python
internal_contours = np.zeros(img.shape)

# Get the internal contours
for i in range(len(contours)):
    if(hierarchy[0][i][3] != -1):
        cv2.drawContours(internal_contours, contours, i, 255, 2)
        
disp(internal_contours)
```

### 3. Get contours of a single group (group 4)

```python
internal_contours_grp4 = np.zeros(img.shape)

contour_group_number = 4

# Get the internal contours
for i in range(len(contours)):
    if(hierarchy[0][i][3] == contour_group_number):
        cv2.drawContours(internal_contours_grp4, contours, i, 255, 2)
        
disp(internal_contours_grp4)
```

```python
# NOte that the contour gruop number information is available in the hierarchy array as the elements in dimension 3

internal_contours_grp0 = np.zeros(img.shape)

contour_group_number = 0

# Get the internal contours
for i in range(len(contours)):
    if(hierarchy[0][i][3] == contour_group_number):
        cv2.drawContours(internal_contours_grp0, contours, i, 255, 2)
        
disp(internal_contours_grp0)
```
