# Grid Detection

## 1. Square grid (Chessboard)

```python
# Grid patterns are often used to calibrate cameras
# In opencv the board has to exclusively look like a chessboard
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def disp(img):
    fig = plt.figure(figsize = (12, 12))
    ax = fig.add_subplot()
    ax.imshow(img)
```

```python
, flat_chess = cv2.imread('../DATA/flat_chessboard.png')
plt.imshow(flat_chess)
```

```python
img = flat_chess.copy()
found, corners = cv2.findChessboardCorners(flat_chess, (7,7))
if(found):
    cv2.drawChessboardCorners(img, (7,7), corners, found)
    disp(img)
```

## 2. Dot grid

```python
dots = cv2.imread('../DATA/dot_grid.png')
disp(dots)
```

```python
img = dots.copy()
found, corners = cv2.findCirclesGrid(dots, (10,10), cv2.CALIB_CB_SYMMETRIC_GRID)
if(found):
    cv2.drawChessboardCorners(img, (10, 10), corners, found)
    
disp(img)
```
