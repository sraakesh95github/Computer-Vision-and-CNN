## Edge Detection

### 1. Canny Edge Detector

```python
# Steps involved

# Custom blur
# Gaussian filter to smooth
# Intesity gradients of the image
# Non - maximum suppression
# Double Threshold
# Track edge by hysteresis
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/sammy_face.jpg')
# Edge detection does not require color correction
plt.imshow(img)
```

```python
edges = cv2.Canny(img, threshold1 = 127, threshold2 = 127)
plt.imshow(edges)
```
